```python
import numpy as np
import pandas as pd
# Package to split the data into train and test set
from sklearn.model_selection import train_test_split
# Package to convert the english language text to machine understandable numerical format for prediction
from sklearn.feature_extraction.text import TfidfVectorizer
# Package to impor the specfic model to use - Logistic Resgression
from sklearn.linear_model import LogisticRegression
# Package to evaluate the test set data
from sklearn.metrics import accuracy_score
```

Data Collection & Preprocessing

```python
# loading the data from csv to pandas dataframe
raw_mail_data = pd.read_csv('spam_mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
# replacing the null values with null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

Label Encoding - Converting to 0 & 1 the spam and ham category values

```python
mail_data.loc[mail_data['Category']=='spam','Category',] = 0
mail_data.loc[mail_data['Category']=='ham','Category',] = 1
```

Spam - 0
Ham - 1

```python
# Sperating the data as text & labels
X = mail_data['Message']
Y = mail_data['Category']
```

```python
X
```

```python
Y
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
X_train.shape, X_test.shape, X.shape
```

Feature Extraction

```python
# Transform the text data to feature vectors
# Can be used as input to Logistic regression model

feature_extraction = TfidfVectorizer(min_df=1, stop_words = 'english', lowercase = True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert y_train & Y_test as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Train Logistic Regression Model

```python
model = LogisticRegression()
```

```python
# training logistic regression model with the training data
model.fit(X_train_features, Y_train)
```

Evaluating the trained model

```python
# Prediction on training data
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
accuracy_on_training_data
```

```python
# Prediction on test data
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
accuracy_on_test_data
```
